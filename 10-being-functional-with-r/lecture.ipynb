## 內建數值函數

## `abs()`

取絕對值。

````r
print(abs(-5566))
print(abs(5566))
````

## `sqrt()`

取平方根。

````r
print(sqrt(2))
print(sqrt(3))
````

無條件進位。

````r
print(ceiling(sqrt(2)))
print(ceiling(sqrt(3)))
````

## `floor()`

無條件捨去。

````r
print(floor(sqrt(2)))
print(floor(sqrt(3)))
````

## `round()`

四捨五入。

````r
print(round(sqrt(2)))
print(round(sqrt(3)))
````

## `exp()`

\begin{equation}
exp(x) = e^x \\
\text{where } e = 2.71828...
\end{equation}

````r
print(exp(1))
print(exp(2))
````

````r
xx <- 1:10
yy <- exp(xx)
plot(xx, yy, type = "b")
title("Exponential function")
````

## `log()`

以 $e$ 為底的對數函數。

````r
e = exp(1)
ee = exp(2)
print(log(e))
print(log(ee))
````

````r
xx <- exp(1:10)
yy <- log(xx)
plot(xx, yy, type = "b")
title("Logarithmic Function")
````

## `log10()`

以 10 為底的對數函數。

````r
print(log10(10))
print(log10(10**2))
````

````r
xx <- 10^(1:10)
yy <- log10(xx)
plot(xx, yy, type = "b")
title("Logarithmic Function")
````

## 內建文字處理函數

## `unique()`

取獨一值。

````r
cities <- c("New York", "Boston", "Tokyo", "Kyoto", "Taipei")
countries <- c("United States", "United States", "Japan", "Japan", "Taiwan")
print(unique(cities))
print(unique(countries))
````

## `toupper()`

轉換為大寫。

````r
print(toupper("Luke, use the Force!"))
````

## `tolower()`

轉換為小寫。

````r
print(tolower("Luke, use the Force!"))
````

## `nchar()`

計算文字的字元個數，number of characters。

````r
print(length("Luke, use the Force"))
print(nchar("Luke, use the Force"))
````

## `substr()`

擷取部分文字片段。

````r
luke <- "Luke Skywalker"
print(substr(luke, start = 1, stop = 4))
print(substr(luke, start = 6, stop = nchar(luke)))
````

## `grep()`

回傳指定特徵有出現的索引值。

````r
avengers <- c("The Avengers", "Avengers: Age of Ultron", "Avengers: Infinity War", "Avengers: Endgame")
print(grep(pattern = "Avengers", avengers))
print(grep(pattern = "Endgame", avengers))
````

## `sub()`

取代指定特徵。

````r
skywalker <- "Anakin Skywalker"
print(skywalker)
print(sub(pattern = "Anakin", replacement = "Luke", skywalker))
````

## `strsplit()`

依據指定特徵分割文字。

````r
avengers <- c("The Avengers", "Avengers: Age of Ultron", "Avengers: Infinity War", "Avengers: Endgame")
print(strsplit(avengers, split = " "))
````

## `paste()` 與 `paste0()`

連接文字。

````r
print(paste("Avengers:", "Endgame"))
print(paste0("Avengers:", "Endgame"))
````

## `trimws()`

消除前（leading）或後（trailing）的空白。

````r
luke <- "     Luke, use the Force!     "
print(trimws(luke))
print(trimws(luke, which = "left"))
print(trimws(luke, which = "right"))
````

## 自行定義函數的輸入與輸出個數

## 單一輸入

將攝氏溫度轉換為華氏溫度的函數 `celsius_to_fahrenheit()`。

\begin{equation}
Fahrenheit_{(°F)} = Celsius_{(°C)} \times \frac{9}{5} + 32
\end{equation}

````r
celsius_to_fahrenheit <- function(x) {
    return(x*9/5 + 32)
}
print(celsius_to_fahrenheit(100))
````

## 多個輸入

將身高（公分）與體重（公斤）轉換為 BMI 的函數 `get_bmi()`。

\begin{equation}
BMI = \frac{weight_{kg}}{height_{m}^{2}}
\end{equation}

````r
get_bmi <- function(height, weight) {
    return(weight / (height*0.01)**2)
}
print(get_bmi(216, 147))
````

## 參數有預設值的函數

將溫度轉換為華氏或攝氏的函數 `temperature_convert()`。

````r
temperature_convert <- function(x, to_fahrenheit = TRUE) {
    if (to_fahrenheit) {
        return(x*9/5 + 32)
    } else {
        return((x - 32)*5/9)
    }
}
print(temperature_convert(100))
print(temperature_convert(212, to_fahrenheit = FALSE))
````

## 多個輸出的函數

將身高（公分）與體重（公斤）轉換為 BMI 以及描述的函數 `get_bmi_and_label()`。

來源：<https://en.wikipedia.org/wiki/Body_mass_index>

````r
get_bmi_and_label <- function(height, weight) {
    bmi <- weight / (height/100)**2
    if (bmi > 30) {
        label <- "Obese"
    } else if (bmi < 18.5) {
        label <- "Underweight"
    } else if (bmi > 25) {
        label <- "Overweight"
    } else {
        label <- "Normal weight"
    }
    bmi_and_label <- list(
        bmi = bmi,
        bmi_label = label
    )
    return(bmi_and_label)
}
````

````r
print(get_bmi_and_label(216, 147)) # Shaquille O'Neal
print(get_bmi_and_label(203, 113)) # LeBron James
````

## 向量化函數

## 什麼是向量化函數？

利用函數型函數（Functional functions）將指定函數**向量化**至某個資料結構的技巧。

````r
# 將一個 list 中的每個數字都平方
my_list <- list(
    11,
    12,
    13,
    14,
    15
)
print(my_list)
````

## 老實說，我們會有個衝動這樣寫

```r
my_list**2
```

## 注意這是一個 `list` 而不是向量

````r
for (i in my_list) {
    print(i**2)
}
````

## 使用 `lapply()` 將 `get_squared()` 函數向量化至 `my_list` 之上

````r
get_squared <- function(x) {
    return(x**2)
}

print(lapply(my_list, FUN = get_squared))
````

## 搭配匿名函數（Anonymous functions）將平方運算向量化至 `my_list` 之上

```r
# Named function
named_function <- function(INPUTS) {
    # sequence of statements
    # sequence of statements
    return(OUTPUTS)
}
# Anonymous function
function(INPUTS) return(OUTPUTS)
```

````r
print(lapply(my_list, FUN = function(x) return(x**2)))
````

## R 常用的函數型函數（Functional functions）

- `lapply()`
- `sapply()`
- `apply()`
- `mapply()`

## `sapply()`

- Simplified apply
- 回傳向量而非 `list`

````r
print(sapply(my_list, FUN = function(x) return(x**2)))
````

## `apply()`

將函數向量化至二維的資料結構（`matrix` 與 `data.frame`）

````r
my_matrix <- matrix(1:12, nrow = 2)
print(my_matrix)
print(apply(my_matrix, MARGIN = 1, FUN = sum))
print(apply(my_matrix, MARGIN = 2, FUN = sum))
````

## `mapply()`

將具有多個輸入的函數向量化

````r
weights <- list(91, 82, 113, 147)
heights <- list(231, 191, 203, 216)
bmis <- mapply(FUN = function(h, w) return(w/(h*0.01)**2), heights, weights)
print(bmis)
````
