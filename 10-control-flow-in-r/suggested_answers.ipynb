## è‡ªè¨‚ä¸€å€‹å‡½æ•¸ `calculate_circle` èƒ½å¤ å°‡è¼¸å…¥çš„åŠå¾‘ä¾ç…§åƒæ•¸ `to` è¨ˆç®—ç‚ºé¢ç©æˆ–å‘¨é•·ï¼Œä½¿ç”¨ $\pi = 3.14$ã€‚

\begin{equation}
Area = \pi r^2 \\
Perimeter = 2 \pi r \\
\text{where} \; \pi = 3.14
\end{equation}

- é æœŸè¼¸å…¥ï¼šä¸€å€‹æ•¸å€¼å‘é‡ã€ä¸€å€‹æ–‡å­—å‘é‡
- é æœŸè¼¸å‡ºï¼šä¸€å€‹æ–‡å­—å‘é‡

```r
calculate_circle <- function(r, to) {
    # calculate_circle(3, to='area')
    # 28.26
    # calculate_circle(3, to='perimeter')
    # 18.84
    # calculate_circle(4, to='area')
    # 50.24
    # calculate_circle(4, to='perimeter')
    # 25.12
    ### BEGIN SOLUTION
    my_pi <- 3.14
    if (to == 'area') {
        return(my_pi*r**2)
    } else {
        return(2*my_pi*r)
    }
    ### END SOLUTION
}
```

## è‡ªè¨‚ä¸€å€‹å‡½æ•¸ `convert_degree` èƒ½å¤ å°‡è¼¸å…¥çš„æº«åº¦åº¦æ•¸ä¾ç…§åƒæ•¸ `from` èˆ‡ `to` åœ¨æ”æ°èˆ‡è¯æ°ä¹‹é–“è½‰æ›ã€‚

\begin{equation}
Fahrenheit_{(Â°F)} = Celsius_{(Â°C)} \times \frac{9}{5} + 32 \\
Celsius_{(Â°C)} =  (Fahrenheit_{(Â°F)} - 32) \times \frac{5}{9}
\end{equation}

- é æœŸè¼¸å…¥ï¼šä¸€å€‹æ•¸å€¼å‘é‡ã€å…©å€‹æ–‡å­—å‘é‡
- é æœŸè¼¸å‡ºï¼šä¸€å€‹æ•¸å€¼å‘é‡

```r
convert_degree <- function(x, from, to) {
    # convert_degree(100, from='Celsius', to='Fahrenheit')
    # 212.0
    # convert_degree(0, from='Celsius', to='Fahrenheit')
    # 32.0
    # convert_degree(212, from='Fahrenheit', to='Celsius')
    # 100.0
    # convert_degree(32, from='Fahrenheit', to='Celsius')
    # 0.0
    ### BEGIN SOLUTION
    if (from == 'Celsius') {
        return(x*9/5 + 32)
    } else {
        return((x - 32)*5/9)
    }
    ### END SOLUTION
}
```

## è‡ªè¨‚ä¸€å€‹å‡½æ•¸ `get_bmi_category` èƒ½å¤ å°‡è¼¸å…¥çš„èº«é«˜ï¼ˆå…¬åˆ†ï¼‰ã€é«”é‡ï¼ˆå…¬æ–¤ï¼‰æ›ç®—ç‚º BMI é¡åˆ¥ã€‚

\begin{equation}
BMI = \frac{weight_{kg}}{height_{m}^2}
\end{equation}

<https://en.wikipedia.org/wiki/Body_mass_index>

- é æœŸè¼¸å…¥ï¼šå…©å€‹æ•¸å€¼å‘é‡
- é æœŸè¼¸å‡ºï¼šä¸€å€‹æ–‡å­—å‘é‡

```r
get_bmi_category <- function(height, weight) {
    # get_bmi_category(170, 50)
    # 'Underweight'
    # get_bmi_category(170, 60)
    # 'Normal'
    # get_bmi_category(206, 113)
    # 'Overweight'
    # get_bmi_category(216, 147)
    # 'Obese'
    ### BEGIN SOLUTION
    bmi <- weight / (height*0.01)**2
    if (bmi > 30) {
        category <- 'Obese'
    } else if (bmi >= 25) {
        category <- 'Overweight'
    } else if (bmi >= 18.5) {
        category <- 'Normal'
    } else {
        category <- 'Underweight'
    }
    return(category)
    ### END SOLUTION
}
```

```r
fizz_buzz <- function(x) {
    # fizz_buzz(3)
    # 'Fizz'
    # fizz_buzz(5)
    # 'Buzz'
    # fizz_buzz(15)
    # 'Fizz Buzz'
    # fizz_buzz(16)
    # 16
    ### BEGIN SOLUTION
    if (x %% 15 == 0) {
        return('Fizz Buzz')
    } else if (x %% 3 == 0) {
        return('Fizz')
    } else if (x %% 5 == 0) {
        return('Buzz')
    } else {
        return(x)
    }
    ### END SOLUTION
}
```

## è‡ªè¨‚ä¸€å€‹å‡½æ•¸ `filter_even_and_sum` èƒ½å°‡ä»‹æ–¼ `a` èˆ‡ `b` ä¹‹é–“çš„å¶æ•¸æŒ‘å‡ºä¾†åŠ ç¸½å¾Œè¼¸å‡ºï¼Œå…¶ä¸­ $a \leq b$ã€‚

- é æœŸè¼¸å…¥ï¼šå…©å€‹ `numeric`
- é æœŸè¼¸å‡ºï¼šä¸€å€‹ `numeric`

```r
filter_even_and_sum <- function(a, b) {
    # filter_even_and_sum(2, 4)
    # 6
    # filter_even_and_sum(1, 5)
    # 6
    # filter_even_and_sum(2, 10)
    # 30
    # filter_even_and_sum(1, 11)
    # 30
    ### BEGIN SOLUTION
    x <- a:b
    evens <- x[x %% 2 == 0]
    return(sum(evens))
    ### END SOLUTION
}
```

## è‡ªè¨‚ä¸€å€‹å‡½æ•¸ `filter_negative_and_sum` èƒ½å°‡ `list` ä¸­çš„è² æ•¸æŒ‘å‡ºä¾†åŠ ç¸½å¾Œè¼¸å‡ºã€‚

- é æœŸè¼¸å…¥ï¼šä¸€å€‹ `list`
- é æœŸè¼¸å‡ºï¼šä¸€å€‹ `numeric`

```r
filter_negative_and_sum <- function(x) {
    # filter_negative_and_sum(list(-5, 5, -6, 6))
    # -11
    # filter_negative_and_sum(list(-5,-5, -6, -6))
    # -22
    # filter_negative_and_sum(list(5, 5, 6, 6))
    # 0
    # filter_negative_and_sum(list(-5, -5, 6, 6))
    # -10
    # filter_negative_and_sum(list(5, 5, -6, -6))
    # -12
    ### BEGIN SOLUTION
    negatives_sum <- 0
    for (item in x) {
        if (item < 0) {
            negatives_sum <- negatives_sum + item
        }
    }
    return(negatives_sum)
    ### END SOLUTION
}
```

## è‡ªè¨‚ä¸€å€‹å‡½æ•¸ `get_factors` èƒ½å¤ è¼¸å‡ºä¸€å€‹æ­£æ•´æ•¸çš„æ‰€æœ‰å› æ•¸ã€‚

- é æœŸè¼¸å…¥ï¼šä¸€å€‹ `numeric`
- é æœŸè¼¸å‡ºï¼šä¸€å€‹ `numeric`

```r
get_factors <- function(x) {
    # get_factors(1)
    # 1
    # get_factors(2)
    # 1 2
    # get_factors(3)
    # 1 3
    # get_factors(4)
    # 1 2 4
    # get_factors(5)
    # 1 5
    ### BEGIN SOLUTION
    possible_factors <- 1:x
    is_factor <- x %% possible_factors == 0
    factors <- possible_factors[is_factor]
    return(factors)
    ### END SOLUTION
}
```

## è‡ªè¨‚ä¸€å€‹å‡½æ•¸ `is_prime` èƒ½å¤ åˆ¤æ–·è¼¸å…¥çš„æ­£æ•´æ•¸æ˜¯å¦ç‚ºè³ªæ•¸ï¼ˆæ°å¥½æœ‰å…©å€‹å› æ•¸ 1 èˆ‡è‡ªèº«çš„æ­£æ•´æ•¸ï¼‰ã€‚

- é æœŸè¼¸å…¥ï¼šä¸€å€‹ `numeric`
- é æœŸè¼¸å‡ºï¼šä¸€å€‹ `logical`

```r
is_prime <- function(x) {
    # is_prime(1)
    # FALSE
    # is_prime(2)
    # TRUE
    # is_prime(3)
    # TRUE
    # is_prime(4)
    # FALSE
    # is_prime(5)
    # TRUE
    ### BEGIN SOLUTION
    factors <- get_factors(x)
    n_factors <- length(factors)
    return(n_factors == 2)
    ### END SOLUTION
}
```

è¨»ï¼šæœ¬é¡Œå¯èƒ½éœ€è¦ä½¿ç”¨åˆ° `strsplit` èˆ‡ `toupper` å‡½æ•¸ã€‚

- é æœŸè¼¸å…¥ï¼šä¸€å€‹ `character`
- é æœŸè¼¸å‡ºï¼šä¸€å€‹ `character`

```r
uppercase_firstname <- function(x) {
    # uppercase_firstname('Luke Skywalker')
    # 'LUKE'
    # uppercase_firstname('Anakin Skywalker')
    # 'ANAKIN'
    # uppercase_firstname(c('Luke Skywalker', 'Anakin Skywalker'))
    # 'LUKE' 'ANAKIN'
    # uppercase_firstname(c('Luke Skywalker', 'Anakin Skywalker', 'Darth Vadar'))
    # 'LUKE' 'ANAKIN' 'DARTH'
    ### BEGIN SOLUTION
    split_names <- strsplit(x, split = " ")
    split_firstnames <- sapply(split_names, FUN = function(x) x[[1]][1])
    return(toupper(split_firstnames))
    ### END SOLUTION
}
```
