## è‡ªè¨‚ä¸€å€‹å‡½æ•¸ `count_negative_integers` èƒ½å›å‚³è¼¸å…¥ `list` ä¹‹ä¸­çš„è² æ•¸æœ‰å¹¾å€‹ã€‚

```r
count_negative_integers <- function(x) {
    # count_negative_integers(list(0, 1, 2, 3, 4, 5))
    # 0
    # count_negative_integers(list(0, -1, -2, -3, -4, -5))
    # 5
    # count_negative_integers(list(0, 5, 5, -6, -6))
    # 2
    # count_negative_integers(list(0, -5, -5, 6, 6))
    # 2
    ### BEGIN SOLUTION
    x_vector <- unlist(x)
    negative_x_vector <- x_vector[x_vector < 0]
    return(length(negative_x_vector))
    ### END SOLUTION
}
```

## è‡ªè¨‚ä¸€å€‹å‡½æ•¸ `find_smallest_int` èƒ½å°‡è¼¸å…¥ `list` ä¹‹ä¸­æœ€å°çš„æ•¸å­—æ‰¾å‡ºä¾†ä¸¦ä¸”å›å‚³ã€‚

- é æœŸè¼¸å…¥ï¼šä¸€å€‹ `list`
- é æœŸè¼¸å‡ºï¼šä¸€å€‹ `numeric`

```r
find_smallest_int <- function(x) {
    # find_smallest_int(list(1, 2, 3, 4, 5))
    # 1
    # find_smallest_int(list(-1, -2, -3, -4, -5))
    # -5
    # find_smallest_int(list(5, 5, 6, 6))
    # 5
    # find_smallest_int(list(-5, -5, -6, -6))
    # -6
    ### BEGIN SOLUTION
    x_vector <- unlist(x)
    return(min(x_vector))
    ### END SOLUTION
}
```

## è‡ªè¨‚ä¸€å€‹å‡½æ•¸ `list_mean` èƒ½è¨ˆç®— `list` ä¹‹ä¸­æ‰€æœ‰æ•¸å­—çš„å¹³å‡å€¼ã€‚

- é æœŸè¼¸å…¥ï¼šä¸€å€‹ `list`

```r
list_mean <- function(x) {
    # list_mean(list(1, 2, 3, 4, 5))
    # 3
    # list_mean(list(-1, -2, -3, -4, -5))
    # -3
    # list_mean(list(5, 5, 6, 6))
    # 5.5
    # list_mean(list(-5, -5, -6, -6))
    # -5.5
    ### BEGIN SOLUTION
    x_vector <- unlist(x)
    return(mean(x_vector))
    ### END SOLUTION
}
```

## è‡ªè¨‚ä¸€å€‹å‡½æ•¸ `count_vowels` èƒ½è¨ˆç®—ä¸€å€‹æ–‡å­—å‘é‡ä¸­çš„æ¯éŸ³å­—æ¯å€‹æ•¸ï¼ˆa, e, i, o, uï¼‰ã€‚

- é æœŸè¼¸å…¥ï¼šä¸€å€‹é•·åº¦ç‚º 1 çš„ `character`
- é æœŸè¼¸å‡ºï¼šä¸€å€‹ `numeric`

```r
count_vowels <- function(x) {
    # count_vowels("Anakin Skywalker")
    # 5
    # count_vowels("Luke Skywalker")
    # 4
    # count_vowels("Darth Vadar")
    # 3
    ### BEGIN SOLUTION
    vowels <- c("a", "e", "i", "o", "u")
    tolower_x <- tolower(x)
    split_char <- strsplit(tolower_x, split="")[[1]]
    n_vowels <- 0
    for (char in split_char) {
        if (char %in% vowels) {
            n_vowels <- n_vowels + 1
        }
    }
    return(n_vowels)
    ### END SOLUTION
}
```

## è‡ªè¨‚ä¸€å€‹å‡½æ•¸ `advanced_count_vowels` èƒ½è¨ˆç®—ä¸€å€‹æ–‡å­—å‘é‡ä¸­çš„æ¯éŸ³å­—æ¯å€‹æ•¸ï¼ˆa, e, i, o, uï¼‰ï¼Œä¸¦ä¸”ä»¥ `list` çš„è³‡æ–™çµæ§‹å›å‚³å„å€‹æ¯éŸ³çš„å€‹æ•¸ã€‚

- é æœŸè¼¸å…¥ï¼šä¸€å€‹é•·åº¦ç‚º 1 çš„ `character`
- é æœŸè¼¸å‡ºï¼šä¸€å€‹ `list`

```r
advanced_count_vowels <- function(x) {
    # advanced_count_vowels("Anakin Skywalker")
    # $a
    # 3
    # $e
    # 1
    # $i
    # 1
    # $o
    # 0
    # $u
    # 0
    # advanced_count_vowels("Luke Skywalker")
    # $a
    # 1
    # $e
    # 2
    # $i
    # 0
    # $o
    # 0
    # $u
    # 1
    # advanced_count_vowels("Darth Vadar")
    # $a
    # 3
    # $e
    # 0
    # $i
    # 0
    # $o
    # 0
    # $u
    # 0
    ### BEGIN SOLUTION
    vowels <- c("a", "e", "i", "o", "u")
    tolower_x <- tolower(x)
    split_char <- strsplit(tolower_x, split="")[[1]]
    n_vowels <- list(
        'a'=0,
        'e'=0,
        'i'=0,
        'o'=0,
        'u'=0
    )
    for (char in split_char) {
        if (char %in% vowels) {
            n_vowels[[char]] <- n_vowels[[char]] + 1
        }
    }
    return(n_vowels)
    ### END SOLUTION
}
```
